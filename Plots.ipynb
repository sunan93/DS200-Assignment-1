```python
import matplotlib.pyplot as plt
import numpy as np
 
```

```python
import pandas as pd
df = pd.io.parsers.read_csv("datafile.csv",index_col=0)
df
```

```python
# total passangers plot
killed_pass = []
injured_pass = []

for i in range(len(df.columns)):
    if i%2==0:
        killed_pass.append(df.iloc[0][i])
    else:
        injured_pass.append(df.iloc[0][i])
killed_pass = np.array(killed_pass, dtype=int)
injured_pass = np.array(injured_pass, dtype=int)
print(killed_pass, injured_pass)
```

```python
import numpy as np
# create data
x = np.array([2002, 2003,2004,2005,2006,2007,2008,2009,2010])
# plot lines
plt.plot(x, killed_pass, label = "killed pass")
plt.plot(x, injured_pass, label = "injured pass")
plt.title("Number of passengers injured vs killed")
plt.xlabel("Year")
plt.ylabel("#Passengers")
plt.legend()
plt.show()
```

```python
killed_serv = []
injured_serv = []

for i in range(len(df.columns)):
    if i%2==0:
        killed_serv.append(df.iloc[4][i])
    else:
        injured_serv.append(df.iloc[4][i])
killed_serv = np.array(killed_serv, dtype=int)
injured_serv = np.array(injured_serv, dtype=int)
print(killed_serv, injured_serv)

x1 = x
x2 = x

y1 = killed_serv
y2 = injured_serv
  
plt.scatter(x1, y1, c ="pink", 
            linewidths = 2, 
            marker ="s", 
            edgecolor ="green", 
            s = 50)
  
plt.scatter(x2, y2, c ="yellow",
            linewidths = 2,
            marker ="^", 
            edgecolor ="red", 
            s = 200)

plt.title("Number of servents injured vs killed")
plt.xlabel("Year")
plt.ylabel("#Servents")
plt.show()
```
